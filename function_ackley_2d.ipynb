```python
from numpy import arange
from numpy import exp
from numpy import sqrt 
from numpy import cos
from numpy import e
from numpy import pi
from numpy import meshgrid
import matplotlib.pyplot as plt

# variable values are: a = 20, b = 0.2 and c = 2π.
# xi ∈ [-32.768, 32.768],

def ackley(x1,x2):
  return -20.0 * exp(-0.2 * sqrt(0.5 * (x1**2 + x2**2)))-exp(0.5 * (cos(2 * 
  pi * x1)+cos(2 * pi * x2))) + e + 200

x_min, x_max = -32.768, 32.768

x1_axis = arange(x_min, x_max , 2.0)
x2_axis = arange(x_min, x_max, 2.0)

x1,x2 = meshgrid(x1_axis, x2_axis)

results = ackley(x1,x2)
figure = plt.figure()
axis = figure.gca(projection ='3d')
axis.plot_surface(x1, x2, results, cmap='jet', shade="false")
plt.show()
```
